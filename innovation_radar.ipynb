```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown
from collections import Counter
import re

# Pengaturan visualisasi
sns.set_style('whitegrid')

# --- 1. MUAT DATA & PRA-PROSES OPTIMAL ---
direktori_dasar = r'C:\Users\WINDOWS 11\Randy\Lomba' 
jalur = {
    'penjualan': os.path.join(direktori_dasar, 'sales.csv'),
    'produk': os.path.join(direktori_dasar, 'products.csv'),
    'ulasan': os.path.join(direktori_dasar, 'reviews.csv'),
    'pemasaran': os.path.join(direktori_dasar, 'marketing.csv')
}

def muat_csv_aman(p):
    if os.path.exists(p):
        try:
            return pd.read_csv(p, encoding='utf-8')
        except Exception:
            return pd.DataFrame()
    return pd.DataFrame()

# Muat dataframes
data_penjualan = muat_csv_aman(jalur['penjualan'])
data_produk = muat_csv_aman(jalur['produk'])
data_ulasan = muat_csv_aman(jalur['ulasan'])
data_pemasaran = muat_csv_aman(jalur['pemasaran'])

KOLOM_MAP = {
    'tanggal_transaksi': ['transaction_date', 'date'],
    'harga_total': ['total_price', 'revenue'],
    'unit_terjual': ['units_sold', 'units', 'quantity'],
    'persen_diskon': ['discount_pct', 'discount'],
    'tanggal_peluncuran': ['launch_date', 'date_launch'],
    'biaya_idr': ['spend_idr', 'spend'],
    'tingkat_engagement': ['engagement_rate', 'engagement']
}

def normalisasi_kolom_cepat(df, map_dict):
    df.columns = df.columns.str.lower()
    for target, alts in map_dict.items():
        for alt in alts:
            if alt in df.columns and target not in df.columns:
                df.rename(columns={alt: target}, inplace=True)
                break

normalisasi_kolom_cepat(data_penjualan, KOLOM_MAP)
normalisasi_kolom_cepat(data_produk, KOLOM_MAP)
normalisasi_kolom_cepat(data_pemasaran, KOLOM_MAP)

# Konversi kolom tanggal (Optimized)
for col, df in [('tanggal_transaksi', data_penjualan), 
                ('tanggal_peluncuran', data_produk), 
                ('tanggal_transaksi', data_pemasaran)]:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')


display(Markdown(f"**Bentuk data** ‚Äî penjualan: {data_penjualan.shape}, produk: {data_produk.shape}, ulasan: {data_ulasan.shape}, pemasaran: {data_pemasaran.shape}"))

# --- 2. HITUNG METRIK & PERTUMBUHAN ---
if 'tanggal_transaksi' in data_penjualan.columns:
    data_penjualan = data_penjualan.copy()
    data_penjualan['tahun_bulan'] = data_penjualan['tanggal_transaksi'].dt.to_period('M')
else:
    data_penjualan['tahun_bulan'] = pd.NaT

# Hitung harga_total jika hilang (sama, sudah cukup baik)
if 'harga_total' not in data_penjualan.columns or data_penjualan['harga_total'].isnull().all():
    if 'revenue' in data_penjualan.columns: data_penjualan['harga_total'] = data_penjualan['revenue']
    elif {'unit_terjual','avg_price'}.issubset(data_penjualan.columns):
        data_penjualan['harga_total'] = data_penjualan['unit_terjual'] * data_penjualan['avg_price']
    else: data_penjualan['harga_total'] = 0.0
if 'unit_terjual' not in data_penjualan.columns:
    data_penjualan['unit_terjual'] = 0

# Agregasi bulanan
bulanan = data_penjualan.groupby(['product_id','tahun_bulan']).agg({'unit_terjual':'sum','harga_total':'sum'}).reset_index()
bulanan.columns = ['product_id','bulan','unit','pendapatan']
bulanan = bulanan.sort_values(['product_id','bulan'])
bulanan['pertumbuhan_pendapatan'] = bulanan.groupby('product_id')['pendapatan'].pct_change()*100

# Agregasi pertumbuhan produk (sama, sudah optimal)
pertumbuhan_produk = bulanan.groupby('product_id').agg({'pendapatan':['sum','mean'],'pertumbuhan_pendapatan':'mean','unit':['sum','mean']}).reset_index()
pertumbuhan_produk.columns = ['product_id','total_pendapatan','rata_rata_pendapatan_bulanan','rata_rata_pertumbuhan_pendapatan','total_unit','rata_rata_unit_bulanan']

# Gabung dengan data pemasaran & hitung ROMS (sama, sudah optimal)
agregasi_pemasaran = data_pemasaran.groupby('product_id').agg({'biaya_idr':'sum','tingkat_engagement':'mean'}).reset_index() if not data_pemasaran.empty else pd.DataFrame(columns=['product_id','biaya_idr','tingkat_engagement'])
agregasi_pemasaran.columns = ['product_id','total_biaya','rata_rata_engagement']
pertumbuhan_produk = pertumbuhan_produk.merge(agregasi_pemasaran, on='product_id', how='left')
pertumbuhan_produk['total_biaya'] = pertumbuhan_produk['total_biaya'].fillna(1.0) # Hindari pembagian dengan nol
pertumbuhan_produk['ROMS'] = pertumbuhan_produk['total_pendapatan'] / pertumbuhan_produk['total_biaya']
pertumbuhan_produk['rata_rata_engagement'] = pertumbuhan_produk['rata_rata_engagement'].fillna(0)

# Peringkat & Kategori
pertumbuhan_produk['peringkat_pertumbuhan'] = pertumbuhan_produk['rata_rata_pertumbuhan_pendapatan'].rank(method='dense', ascending=False)
pertumbuhan_produk['peringkat_roms'] = pertumbuhan_produk['ROMS'].rank(method='dense', ascending=False)
pertumbuhan_produk['peringkat_gabungan'] = (pertumbuhan_produk['peringkat_pertumbuhan'] + pertumbuhan_produk['peringkat_roms']).rank(method='dense')
lima_teratas = pertumbuhan_produk.nlargest(5, 'peringkat_gabungan') if not pertumbuhan_produk.empty else pertumbuhan_produk.head(0)
tiga_optimal = pertumbuhan_produk.nsmallest(3, 'peringkat_gabungan')['product_id'].tolist()

display(Markdown('**5 Kandidat Teratas (berdasarkan peringkat gabungan)**'))
display(pertumbuhan_produk.sort_values('peringkat_gabungan').head(10).reset_index(drop=True))

def region_teratas_untuk(pid):
    if 'region' in data_penjualan.columns:
        reg = data_penjualan[data_penjualan['product_id']==pid].groupby('region')['harga_total'].sum().reset_index()
        if len(reg): return reg.sort_values('harga_total', ascending=False).iloc[0]['region']
    return 'N/A'

def olah_kata_kunci(pid):
    sw = {'yang','ini','dan','atau','untuk','dari','ke','di','dengan','tidak','sangat','sudah','aja','pun','juga','bisa','saya','anda','nya','produk','barang','dan'}
    if not data_ulasan.empty:
        teks = data_ulasan[data_ulasan['product_id']==pid]['comment'].fillna('').str.lower()
        token = re.sub(r'[^a-z0-9\s]','', ' '.join(teks)).split()
        kata_kunci = [w for w in token if len(w)>2 and w not in sw]
        return [k for k,_ in Counter(kata_kunci).most_common(5)]
    return []

# --- 3. RINGKASAN EKSEKUTIF & TABEL 5W+1H ---
baris_ringkasan = []
for pid in tiga_optimal:
    g = pertumbuhan_produk[pertumbuhan_produk['product_id']==pid].iloc[0]
    wilayah = region_teratas_untuk(pid)
    kata_kunci = olah_kata_kunci(pid)
    baris_ringkasan.append({
        'Produk': pid,
        'Apa (Pertumbuhan %)' : f"{g['rata_rata_pertumbuhan_pendapatan']:.1f}",
        'Di Mana (Wilayah Teratas)': wilayah,
        'Kapan (Petunjuk Adopsi Awal)': 'Lihat plot KAPAN',
        'Siapa (Kata Kunci)': ', '.join(kata_kunci) if kata_kunci else 'N/A',
        'Mengapa (Engagement/Diskon)': f"Eng:{g['rata_rata_engagement']:.1f} | ROMS:{g['ROMS']:.2f}x",
        'Bagaimana (Saluran Dominan)': 'Lihat plot BAGAIMANA',
        'Efisiensi (ROMS)': f"{g['ROMS']:.2f}x"
    })

df_ringkasan = pd.DataFrame(baris_ringkasan)
display(Markdown('# RINGKASAN EKSEKUTIF ‚Äî 3 Produk Paling Optimal'))
for i, r in df_ringkasan.iterrows():
    md = f"**RANK #{i+1}: {r['Produk']}** ¬†\n- APA: Rata-rata pertumbuhan bulanan **{r['Apa (Pertumbuhan %)']}%** ¬†\n- DI MANA: Wilayah teratas **{r['Di Mana (Wilayah Teratas)']}** ¬†\n- MENGAPA: {r['Mengapa (Engagement/Diskon)']} ¬†\n- EFISIENSI: **{r['Efisiensi (ROMS)']}** ¬†\n"
    display(Markdown(md))
display(Markdown('## Tabel 5W+1H (Top 3)'))
display(df_ringkasan.reset_index(drop=True))

# --- 4. INTERPRETASI RINGKAS & REKOMENDASI YANG DIREVISI ---
display(Markdown('## Interpretasi Ringkas & Rekomendasi Eksekutif'))
display(Markdown('### üåü Ringkasan *Insight* Data'))
display(Markdown('''
* **Top-3 Produk Optimal (Berlaku untuk Produk Peringkat 1-3):** Produk ini (misalnya, **P01, P02, P03**) secara konsisten menunjukkan kombinasi **Pertumbuhan Kuat** dan **Efisiensi Pemasaran Tinggi (ROMS)**. Mereka adalah kandidat utama untuk **Ekspansi Pasar (Scale)**. Strategi utamanya adalah duplikasi dan percepatan.
* **Efisiensi Terhadap Pertumbuhan:**
    * **ROMS Tinggi, Pertumbuhan Lambat (Potensi *Cash Cow*):** Produk yang sangat efisien tetapi pertumbuhannya datar memberikan **peluang *testing* risiko rendah** pada distribusi atau promosi. Dana dari produk ini harus digunakan untuk mendanai produk *optimal* atau *tanda tanya*.
    * **Pertumbuhan Kuat, ROMS Rendah (Potensi *Question Mark*):** Perlu dilakukan analisis mendalam terhadap **kanal pemasaran** dan **materi promosi** produk ini. Ini adalah sinyal untuk perbaikan efektivitas kampanye (optimasi target atau pemotongan saluran yang tidak efektif).
* **Kualitas & *Insight* Pelanggan:** Isu yang muncul berulang kali di **Kata Kunci Ulasan (SIAPA)**‚Äîseperti komplain **kemasan**, **keterlambatan**, atau **fungsi**‚Äîberdampak langsung pada Sentimen dan retensi. Perbaikan di area ini dapat meningkatkan konversi dan ROMS secara organik.
'''))
display(Markdown('### üó∫Ô∏è Panduan Membaca Visualisasi 5W+1H'))
display(Markdown('''
| Bagian Plot | Pertanyaan (W/H) | Cara Membaca | Aksi yang Disarankan |
| :--- | :--- | :--- | :--- |
| **Tren Pendapatan Bulanan** | **APA** | Lihat kemiringan garis: Garis yang curam dan stabil menunjukkan momentum pertumbuhan yang sehat. | Identifikasi bulan-bulan puncak dan terendah untuk sinkronisasi inventaris dan promosi. |
| **Distribusi Regional** | **DI MANA** | Batang terpanjang mewakili wilayah dominan. Bar Chart menunjukkan potensi pasar yang belum tersentuh. | Prioritaskan alokasi *marketing spend* (biaya) dan distribusi ke wilayah teratas (dominan). |
| **Adopsi Awal** | **KAPAN** | Laju penjualan (unit terjual) dalam **180 Hari Pertama**. Garis merah di Hari ke-90 sering menjadi titik evaluasi. | Bandingkan adopsi Top-3: Jika salah satu memiliki adopsi awal lambat tetapi pertumbuhan tinggi, **strategi *long-tail* efektif**. |
| **Kata Kunci Ulasan** | **SIAPA** | Kata yang paling sering muncul mencerminkan **kekuatan** atau **kelemahan** produk di mata pelanggan. | Jika kata kunci negatif mendominasi (misalnya "rusak", "lama"), jadikan **perbaikan produk/layanan** sebagai KPI utama. |
| **Engagement & Diskon** | **MENGAPA** | Lihat sebaran Engagement vs Pendapatan: Produk di kanan atas (Engagement & Pendapatan tinggi) adalah produk yang paling disukai. | Analisis: Apakah diskon yang tinggi menghasilkan pertumbuhan yang sebanding? Jika ya, diskon mungkin menjadi *driver* utama. |
| **Distribusi Saluran** | **BAGAIMANA** | Warna yang mendominasi pada Top-5 menunjukkan **Saluran Dominan** (*Channel MVP*). | Alokasikan 80% biaya iklan ke saluran MVP untuk Top-3 produk optimal. |
'''))
display(Markdown('### üöÄ Rekomendasi Eksekutif Singkat'))
display(Markdown('''
1. **Alokasi Anggaran Terfokus:** Prioritaskan 1) **Produk Optimal (Peringkat 1-3)** untuk ditingkatkan/scale, 2) Produk *Question Mark* (Pertumbuhan Tinggi/ROMS Rendah) untuk eksperimen **optimasi biaya terbatas**, dan 3) Produk Efisiensi Rendah (Anjing) untuk dipantau ketat / dipertimbangkan pemotongan biaya.
2. **Perbaikan Berbasis Ulasan:** Tinjau hasil plot **SIAPA** (Kata Kunci Ulasan). Lakukan perbaikan cepat pada isu-isu seperti *packaging* atau logistik, karena ini berdampak langsung pada **Sentimen Pelanggan** dan mengurangi potensi biaya retur/komplain.
3. **Eksperimen Pemasaran Cepat:** Jalankan 2‚Äì3 A/B *test* selama 4 minggu (campuran saluran, pesan promosi, promo kecil) yang berfokus pada produk *Question Mark* untuk mengukur dan meningkatkan ROMS secara terukur.
4. **Monitoring Proaktif:** Buat daftar pantauan harian/mingguan untuk **Top-3** dan siapkan notifikasi (*alert*) jika terjadi penurunan ROMS **>20%** atau lonjakan ulasan negatif.
'''))
display(Markdown('Catatan: Angka-angka presisi (pertumbuhan%, ROMS, wilayah, persentase saluran) tersedia di tabel-tabel sebelumnya ‚Äî gunakan nilai tersebut untuk menentukan ukuran anggaran dan target KPI per eksperimen.'))

# --- 5. VISUALISASI ---
# 5.1 APA - Tren pendapatan 5 teratas
display(Markdown('### APA ‚Äî Tren Pendapatan Bulanan (5 Kandidat Teratas)'))
top5_ids = pertumbuhan_produk.nsmallest(5,'peringkat_gabungan')['product_id'].tolist()
fig, ax = plt.subplots(figsize=(10,5))
for pid in top5_ids:
    d = bulanan[bulanan['product_id']==pid]
    if d.empty: continue
    ax.plot(d['bulan'].astype(str), d['pendapatan'], marker='o', label=pid)
ax.set_title('Pendapatan Bulanan ‚Äî Kandidat Teratas'); ax.set_xlabel('Bulan'); ax.set_ylabel('Pendapatan'); plt.xticks(rotation=45); ax.legend(); plt.tight_layout(); plt.show()

# 5.2 DI MANA - Distribusi pendapatan regional
display(Markdown('### DI MANA ‚Äî Distribusi Pendapatan Regional untuk 3 Produk Optimal'))
if 'region' in data_penjualan.columns and tiga_optimal:
    fig, axes = plt.subplots(1, len(tiga_optimal), figsize=(12,4))
    for i,pid in enumerate(tiga_optimal):
        data = data_penjualan[data_penjualan['product_id']==pid].groupby('region')['harga_total'].sum().sort_values(ascending=True)
        ax = axes[i] if len(tiga_optimal)>1 else axes
        data.plot(kind='barh', ax=ax); ax.set_title(pid); ax.set_xlabel('Pendapatan')
    plt.tight_layout(); plt.show()
else:
    display(Markdown('Data wilayah tidak tersedia'))

# 5.3 KAPAN - Adopsi awal
display(Markdown('### KAPAN ‚Äî Adopsi Awal (Unit vs Hari Sejak Peluncuran, 180 Hari Pertama)'))
if 'tanggal_peluncuran' in data_produk.columns and 'tanggal_transaksi' in data_penjualan.columns:
    s = data_penjualan.merge(data_produk[['product_id','tanggal_peluncuran']], on='product_id', how='left')
    s['hari_sejak_peluncuran'] = (s['tanggal_transaksi'] - s['tanggal_peluncuran']).dt.days
    p = s[(s['hari_sejak_peluncuran'].between(0,180))].groupby(['product_id','hari_sejak_peluncuran'])['unit_terjual'].sum().reset_index()
    fig, ax = plt.subplots(figsize=(10,5))
    for pid in p['product_id'].unique():
        d = p[p['product_id']==pid]
        alpha = 0.8 if pid in tiga_optimal else 0.15
        ax.scatter(d['hari_sejak_peluncuran'], d['unit_terjual'], s=20, alpha=alpha, label=pid if pid in tiga_optimal else '')
    ax.set_xlabel('Hari sejak peluncuran'); ax.set_ylabel('Unit terjual'); ax.axvline(90, color='red', linestyle=':'); plt.legend(); plt.tight_layout(); plt.show()
else:
    display(Markdown('Tanggal peluncuran atau tanggal transaksi hilang ‚Äî tidak dapat membuat plot KAPAN'))

# 5.4 SIAPA - Kata kunci
display(Markdown('### SIAPA ‚Äî Kata Kunci Teratas (Ulasan) untuk Top 3'))
if not data_ulasan.empty and tiga_optimal:
    fig, axes = plt.subplots(1, len(tiga_optimal), figsize=(12,4))
    sw = {'yang','ini','dan','atau','untuk','dari','ke','di','dengan','tidak','sangat','sudah','aja','pun','juga','bisa','saya','anda','nya','produk','barang','dan'}
    for i,pid in enumerate(tiga_optimal):
        teks = data_ulasan[data_ulasan['product_id']==pid]['comment'].fillna('').str.lower()
        token=[]
        token = olah_kata_kunci(pid)
        kata_kunci = Counter(token).most_common(10)
        if kata_kunci:
            kata, hitungan = zip(*kata_kunci)
            ax = axes[i] if len(tiga_optimal)>1 else axes
            ax.barh(kata, hitungan); ax.set_title(pid)
        else:
            ax = axes[i] if len(tiga_optimal)>1 else axes; ax.text(0.5,0.5,'Tidak ada kata kunci'); ax.set_axis_off()
    plt.tight_layout(); plt.show()
else:
    display(Markdown('Data ulasan tidak tersedia untuk plot SIAPA'))

# 5.5 MENGAPA - Engagement & Diskon
display(Markdown('### MENGAPA ‚Äî Perbandingan Engagement vs Pendapatan dan Diskon'))
if not pertumbuhan_produk.empty:
    pm = pertumbuhan_produk.copy()
    fig, axes = plt.subplots(1,2, figsize=(12,5))
    sc = axes[0].scatter(pm['rata_rata_engagement'], pm['total_pendapatan'], s=(pm['total_pendapatan']/1e8).clip(lower=10), c=pm['ROMS'], cmap='RdYlGn'); plt.colorbar(sc, ax=axes[0], label='ROMS'); axes[0].set_xlabel('Rata-rata Engagement'); axes[0].set_ylabel('Total Pendapatan')
    if 'persen_diskon' in data_penjualan.columns or 'avg_discount' in data_produk.columns:
        if 'persen_diskon' in data_penjualan.columns:
            diskon = data_penjualan.groupby('product_id')['persen_diskon'].mean().reset_index().rename(columns={'persen_diskon':'rata_rata_diskon'})
        else:
            diskon = data_produk[['product_id','avg_discount']].copy().rename(columns={'avg_discount':'rata_rata_diskon'})
        pm2 = pm.merge(diskon, on='product_id', how='left')
        pm2['kategori'] = pm2['product_id'].apply(lambda x: 'Radar Inovasi (Optimal)' if x in tiga_optimal else 'Lainnya')
        sns.boxplot(data=pm2, x='kategori', y='rata_rata_diskon', ax=axes[1])
        axes[1].set_xlabel('Kategori Produk'); axes[1].set_ylabel('Rata-rata Diskon (%)')
    else:
        axes[1].text(0.5,0.5,'Data diskon tidak tersedia', ha='center'); axes[1].set_axis_off()
    plt.tight_layout(); plt.show()
else:
    display(Markdown('Data tidak mencukupi untuk plot MENGAPA'))

# 5.6 BAGAIMANA - Distribusi saluran
display(Markdown('### BAGAIMANA ‚Äî Distribusi Saluran (Top 5 vs Semua Produk)'))
if 'channel' in data_penjualan.columns:
    distribusi_saluran = data_penjualan.groupby(['product_id','channel'])['unit_terjual'].sum().reset_index()
    distribusi_saluran['total'] = distribusi_saluran.groupby('product_id')['unit_terjual'].transform('sum')
    distribusi_saluran['persen'] = distribusi_saluran['unit_terjual']/distribusi_saluran['total']*100
    topk = pertumbuhan_produk.nsmallest(5,'peringkat_gabungan')['product_id'].tolist()
    pivot_saluran = distribusi_saluran[distribusi_saluran['product_id'].isin(topk)].pivot(index='product_id', columns='channel', values='persen').fillna(0)
    fig, ax = plt.subplots(figsize=(10,4))
    pivot_saluran.plot(kind='bar', stacked=True, ax=ax); ax.set_ylabel('Persentase') ; plt.tight_layout(); plt.show()
else:
    display(Markdown('Data saluran tidak tersedia untuk plot BAGAIMANA'))
```
